```python
import json
from datetime import datetime
import os

path = "page_post"
pages = os.listdir(path)

min_time = datetime(2025, 1, 1, 0, 0, 0)

fields = ["creation_time", "text", "permalink", "comment_count", "reaction_count", "share_count"]
for page in pages:
    posts = os.listdir(os.path.join(path, page))
    datas = []
    for post in posts:
        try:
            with open(f"{path}/{page}/{post}/{post}.json", "r") as f:
                data = json.load(f)
        except Exception as e:
            print(f"Error: {page}/{post}")
            continue

        try:
            for field in fields:
                assert data[field] is not None
        except Exception as e:
            if field == "text" and "event" in data["permalink"]:
                pass
            else:
                print(f"Error: {page}/{post} - {field}")
            continue

        datas.append(data['creation_time'])

    min_time = datetime.fromtimestamp(min(datas))
    max_time = datetime.fromtimestamp(max(datas))
    post_count = len(datas)
    print(f"{page}")
    print(f"\tmin_time: {min_time} ", "✅" if min_time <= datetime(2025, 1, 1, 0, 0, 0) else "❌")
    print(f"\tmax_time: {max_time}")
    print(f"\tpost count: {post_count}")
    print()
```

```python
from pathlib import Path
import shutil

source_root = Path("page_post")
destination_root = Path("page_post_fb")

for json_file in source_root.rglob("*.json"):
    # Create the same relative path in destination
    relative_path = json_file.relative_to(source_root)
    destination_file = destination_root / relative_path

    # Create parent folders if they don't exist
    destination_file.parent.mkdir(parents=True, exist_ok=True)

    # Copy the json file
    shutil.copy2(json_file, destination_file)

print("Done copying JSON files with original structure.")
```
